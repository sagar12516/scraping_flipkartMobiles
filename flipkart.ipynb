```python
def get_data():
    title = []
    ratings = []
    reviews = []
    description = []
    price = []
    assured = []
    listings = soup.find_all('div',class_='_1-2Iqu row')
    for i in listings:
        title.append(i.div(class_='_3wU53n')[0].text)
        r=i.find_all('div',class_='hGSR34')
        if len(r)!=0:
            ratings.append(r[0].text)
        else:
            ratings.append(np.nan)
        try:
            a1 = i.find_all('span',class_='_38sUEc')[0]
            a = list(a1.span)[-1].text
            b =re.findall(r'[0-9]',a)
            c = ''.join(b)
            if len(c)!=0:
                reviews.append(c)
            else:
                reviews.append(np.nan)
        except:
            reviews.append(np.nan)
        d=i.find_all('ul',class_='vFw0gD')
        if len(d)!=0:
            description.append(d[0].text)
        else:
            description.append(np.nan)
        p=i.find_all('div',class_='_1vC4OE _2rQ-NK')
        if len(p)!=0:
            price.append(p[0].text)
        else:
            price.append(np.nan)
        f = i.find_all('div',class_='_3n6o0t')
        if len(f)!=0:
            assured.append('Yes')
        else:
            assured.append('No')
    table = pd.DataFrame({'Title':title,'Rating':ratings,'Reviews':reviews,'Description':description,'Price':price,'Assured':assured})
    return table
```

```python
import time
tables = []
btn = browser.find_elements_by_class_name('_3fVaIS')
flag = True
tables.append(get_data())
btn[0].click()
html = browser.page_source
soup = BeautifulSoup(html,'html.parser')
while flag:
    html = browser.page_source
    soup = BeautifulSoup(html,'html.parser')
    btn = browser.find_elements_by_class_name('_3fVaIS')
    table = get_data()
    tables.append(table)
    if len(btn)==2 and soup.find_all('a',class_='_3fVaIS')[1].span.text.strip()=='Next':
        try:
            time.sleep(0.5)
            btn[1].click()
        except:
            time.sleep(20)
            btn = browser.find_elements_by_class_name('_3fVaIS')
            if len(btn)==2 and  soup.find_all('a',class_='_3fVaIS')[1].span.text.strip()=='Next':
                btn[1].click()
            else:
                flag=False
    else:
        flag=False
```
